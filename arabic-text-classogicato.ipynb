# Data cleaing 

```python
#remove some marks 
# sampl=r'#.][!XR'

# for letter in r'#.][!XR':
#     clean_data['text'] = clean_data['text'].astype(str).str.replace(letter,'',regex=True)
for letter in r'#.][!XR':
    clean_data['text'] = clean_data['text'].astype(str).str.replace(re.escape(letter), '', regex=True)
```

```python
english_punc = string.punctuation
arabic_punc = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''
all_punc= english_punc + arabic_punc 
all_punc
```

```python
def remove_punc (text): 
    translator = str.maketrans('','',all_punc)
    return text.translate(translator)
```

```python
import re
def remove_repeating_char(text):
    return re.sub(r'(.)\1+',r'\1',text)
# هيبحث عن . الكثيرة في النص ز يسنبديها ب . واحدة فقط 
# .................... -> .
```

```python
def normalize_arabic (text):
    text = re.sub("[إأآا]",  "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ة", "ه",text)
    text = re.sub("گ", "ك",text)
    return text
############## گ ف النص هنستبدل ال ك ب 
```

```python
stemmer = ISRIStemmer()
```

```python
def processDoc(doc, stemmer):
    doc = remove_punc(doc)
    doc = re.sub(r'@[^\s]+', ' ', doc)
    doc = re.sub(r'_', ' ', doc)
    doc = re.sub(r'\n', ' ', doc)
    doc = re.sub(r'[a-z,A-Z]', '', doc)
    doc = re.sub(r'\d', '', doc)
    doc = re.sub('((www\.[^\s]+)|(https?://[^\s]+))',' ',doc)
    doc = normalize_arabic ( doc)
    doc = remove_repeating_char (doc)
    
    doc = stemmer.stem(doc)
    
    return doc
```

```python
l=processDoc(clean_data.iloc[0,0],stemmer)
```

```python
clean_data['text']= clean_data['text'].apply(lambda x : processDoc(x,stemmer))
```

# Text Tokenization 

```python
#to be used with input after training 
def clen (inp):
    text =processDoc(inp, stemmer)
    tokenizer= RegexpTokenizer(r'\w+')
    tokenized_text=tokenizer.tokenize(text) 
    stopwords_list=stopwords.words("arabic")
    remove_stop_words=[i for i in tokenized_text if not i in stopwords_list  ]
    return remove_stop_words
    
    
```

```python
tokenizer= RegexpTokenizer(r'\w+')
clean_data ['text']=clean_data ['text'].apply(tokenizer.tokenize)
```

```python
clean_data.iloc[3787,0]
```

```python
stopwords_list=stopwords.words("arabic")
clean_data['text']= clean_data['text'].apply(lambda x: [item for item in x if item not in stopwords_list ])
```

# Data spliting 

```python
x=clean_data["text"]
y=clean_data["targe"]

```

```python
X_train , X_test , y_train,y_test = train_test_split(x,y,test_size=0.25,random_state =42,shuffle=True )
```

```python
word_vec=TfidfVectorizer(sublinear_tf=True,strip_accents ='unicode',analyzer ='word',ngram_range=(1,1), max_features = 10000)
```

```python
SVM_model= Pipeline([('tfidf',word_vec),("clf",SVC())])
```

```python
SVM_model.fit(X_train.astype('str'),y_train)
```

```python
dump(SVM_model,'SVM_model.joblib')
```

```python
## to load this pretrained Model
import joblib

loaded_model = joblib.load('SVM_model.joblib')
loaded_model
```

```python
y_pred = loaded_model.predict(X_test.astype('str'))
# ax = sns.histplot(y_pred, discrete=True)
# ax.set_xticklabels(["", "culture", "diverse", "economy", "politics", "sport"])
```

```python
model_accuracy = accuracy_score(y_test,y_pred)
```
